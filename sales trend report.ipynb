### Analyze amazon sales data to identify trends over time and provide recommendations to improve sales performance.



## 1. Data Import and Setup
The initial step involves importing the necessary libraries and loading the dataset into a Pandas DataFrame for analysis. We'll then display the first 5 rows to ensure the data has been successfully imported.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

Amazon_df = pd.read_csv(r'C:\Users\Mohsu\Desktop\python workspace\Amazon Data\amazon.csv')
print("this is the first 5 row of Amazon_df datafram :")
Amazon_df.head()
```

```python
print("\n this is the Amzon_df  information : ")
Amazon_df.info()
```

## **2. Data Cleaning**

###  handle missing values

Convert price-related columns to numeric and handle missing values

```python
# clean the data from the price-related columns 
chr_to_remove = ['₹',',','%']
col_to_clean = ['discounted_price','actual_price','discount_percentage','rating', 'rating_count' ]

for item in chr_to_remove:
    for cols in col_to_clean :
        Amazon_df[cols] = Amazon_df[cols].str.replace(item , '')

# convert the price-related columns into numeric 
for col in col_to_clean:
    Amazon_df[col] = pd.to_numeric(Amazon_df[col], errors='coerce')

# fill the missing values

for col in col_to_clean:
    Amazon_df[col].fillna(Amazon_df[col].mean(), inplace=True)

print('\n this is the data after clean it :')
Amazon_df.head()
```

```python
# check the data after Convert columns to numeric
print('\n this is the data information :')
Amazon_df.info()
```

```python
# check the missing values 
Amazon_df_missing = Amazon_df.isnull().sum()
print('\n this is the missing  data:')

Amazon_df_missing
```

```python
# check the unique value in category 
Amazon_df['category'].unique()
```

```python
Amazon_df['Main_Category'] = Amazon_df['category'].str.split('|').str[0]
Amazon_df['Subcategory'] = Amazon_df['category'].str.split('|').str[1]
Amazon_df['Detailed_Subcategory'] = Amazon_df['category'].str.split('|').str[2]
Amazon_df['Product_Group'] = Amazon_df['category'].str.split('|').str[3]
Amazon_df['Specific_Item'] = Amazon_df['category'].str.split('|').str[4]

Amazon_df.head()
```

```python
Amazon_df.drop('category', axis=1 , inplace= True)
Amazon_df.head()
```

## 3. Creating Sales Metrics 

 Adding  a revenue column and Calculate overall category performance

```python
Amazon_df["revenue"] = Amazon_df["actual_price"] - Amazon_df["discounted_price"]

category_performance = Amazon_df.groupby("Main_Category").agg({
    "revenue": "sum",
    "actual_price": "sum",
    "discounted_price": "mean"
}).sort_values(by="revenue", ascending=False)

print("\n the  Main_Categorys :")
category_performance
```

```python
top_products = Amazon_df.nlargest(5, "revenue")
print("This is the top 5 products:")
print(top_products[['product_name', 'revenue']])
```

## 4.Visualizations


 Bar Chart for Revenue by Category

```python

plt.figure(figsize=(10, 6))

# Plot Revenue
sns.barplot(x='Main_Category', y='revenue', data=category_performance)

# Title and Labels
plt.title('Revenue by Category')
plt.xlabel('Main_Category')
plt.ylabel('Revenue')
plt.xticks(rotation=25, fontsize=10)  

# Show the plot
plt.tight_layout()
plt.show()
```

Impact of discount percentage on revenue

```python

plt.figure(figsize=(10, 5))
sns.scatterplot(data=Amazon_df, x="discount_percentage", y="revenue", hue="Main_Category")
plt.title("Impact of Discount Percentage on Revenue")
plt.show()
```

### Key Insights:
1. **Category Performance**: Electronics generated the highest revenue, followed by Home Appliances. Discounts played a significant role in driving sales for these categories.
2. **Discount Effectiveness**: Discounts between 30%-50% had the most significant impact on sales, as seen in the scatterplot.


### Recommendations:
- Increase marketing efforts for top-performing categories and products.
- Experiment with targeted discounts between 30%-50% to maximize revenue.
- Analyze low-performing categories for improvement opportunities or consider reducing inventory.
